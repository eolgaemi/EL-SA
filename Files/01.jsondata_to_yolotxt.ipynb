## Json으로된 라벨링 파일 txt로 변경하는 부분(중복이름은 처음만 저장하고 이후에는 저장안함)

```python
import json

def convert_to_yolo(label, img_width, img_height): # json파일 안에 들어있는 내용과 yolo에서 사용되는 txt파일의 값이 다름으로 변환하는 부분
    x = label['x']
    y = label['y']
    w = label['width']
    h = label['height']

    # Normalize the values
    x_center = (x + w / 2) / img_width
    y_center = (y + h / 2) / img_height
    width = w / img_width
    height = h / img_height

    return x_center, y_center, width, height

def process_json(json_file):
    with open(json_file, 'r') as file:
        data = json.load(file)
    
    img_width = data['metadata']['width']
    img_height = data['metadata']['height']
    
    for frame in data['frames']:
        img_name = frame['image']
        annotations = frame['annotations']
        
        # Create a corresponding .txt file for each image
        txt_filename = img_name.replace('.jpg', '.txt')
        if os.path.exists(os.path.join('C:/Users/USER/Documents/Documents/Documents/Detection/Label/',txt_filename)):
            print(json_file)
            continue
        else:
            with open('C:/Users/USER/Documents/Documents/Documents/Detection/Label/'+txt_filename, 'w') as out_file:
                for annotation in annotations:
                    if not  annotation['category']['attributes']:# 중복되는 값이 나오면 저장하지 않고 넘어가게 됩니다.
                        continue
                    else:
                        category_id =annotation['category']['code']
                        if  category_id == "wheelchair":
                            category_id ="0"
                        x_center, y_center, width, height = convert_to_yolo(annotation['label'], img_width, img_height)
                        out_file.write(f"{category_id} {x_center} {y_center} {width} {height}\n")
for i in range(0,19):
# JSON 파일 경로
    json_file_path = 'C:/Users/USER/Documents/Documents/Documents/Detection/휠체어_17_Labeling/'+file_names[i] # JSON 파일 경로를 정확히 설정하세요.

# JSON 파일을 YOLO 포맷으로 변환
    process_json(json_file_path)
```

## 중복 이미지 데이터 삭제후 폴더에서 꺼내기(중복된 이미지는 print문으로 확인 가능)

```python
import os
import shutil
allpath = 'C:/Users/USER/Documents/Documents/Documents/Detection/휠체어_17_image2/'
file_names = os.listdir(allpath)#years
new_file=''
for i in file_names:
    fpath = os.path.join(allpath, i)  # 파일 경로
    in_fpath = os.listdir(fpath)  # 파일 경로 안의 모든 파일 가져오기
    for filename in in_fpath:
        src = os.path.join(fpath, filename)  # 파일의 전체 경로
        if os.path.exists(os.path.join(allpath,filename)):
            os.remove(src)
            print(src)
        else:
            shutil.move(src, allpath)  # 파일을 allpath로 이동
    os.rmdir(fpath)
```
